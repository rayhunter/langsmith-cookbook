**Simple Chains** simple task

```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

llm = ChatOpenAI(model_name='gpt-3.5-turbo', temperature=0.5)
template = '''Your are an experienced virolist.
Write a few sentences about. the following {virus}.
'''
#Write a few sentences about. the following {virus} in {language}.

prompt = PromptTemplate(
    #input_variables=['virus','language'],
    input_variables=['virus'],
    template=template
)

chain = LLMChain(llm=llm, prompt=prompt)
#output = chain.run({'virus': 'H1N1', 'language': 'Spanish'})
output = chain.run({'virus': 'H1N1'})
```

```python
print(output)
```

**Sequential Chains** complex task

```python
from langchain.chat_models import ChatOpenAI
from langchain.llms import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain, SimpleSequentialChain

llm = OpenAI(model_name='text-davinci-003', temperature=0.7, max_tokens=1024)

prompt = PromptTemplate.from_template(
    imput_variables=['concept'],
    template='''You are an experienced scientist and Python programmer.
    Write a function that implements the concept of {concept}.'''
)
chain = LLMChain(llm=llm, prompt=prompt)

llm2 = ChatOpenAI(model_name='gpt-3.5-turbo', temperature=1.2)
prompt2 = PromptTemplate.from_template(
    imput_variables=['function'],
    template='Given the Python function {function}, describe it as detailed as possible.'
)
chain2 = LLMChain(llm=llm2, prompt=prompt2)

overall_chain = SimpleSequentialChain(chains=[chain, chain2], verbose=True)
output = overall_chain.run('softmax')
```

**Splitting and Embedding Text Using Langchain**

```python
import os
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv(), override=True)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
with open('./ThoughtsOnPortland.txt') as f:
    ThoughtsOnPortland = f.read()
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20,
    length_function=len
)
```

```python
chunks = text_splitter.create_documents([ThoughtsOnPortland])
print(chunks[2])
print(chunks[10].page_content)
print(f'Now you have {len(chunks)}')
```

**Embedding Cost**

```python
def print_embedding_cost(texts):
  import tiktoken
  enc = tiktoken.encoding_for_model('text-embedding-ada-002')
  total_tokens = sum([len(enc.encode(page.page_content)) for page in texts])
  print(f'Total Tokens: {total_tokens}')
  print(f'Embedding Cost in USD: {total_tokens / 1000 * 0.0004:.6f}')

print_embedding_cost(chunks)

```

```python
from langchain.embeddings import OpenAIEmbeddings
embedding = OpenAIEmbeddings()
```

```python
vector = embedding.embed_query(chunks[0].page_content)
print(vector)
```

**Insert Embeddings into Pinecone Index**

```python
import os
import pinecone
from langchain.vectorstores import Pinecone

pinecone.init(api_key=os.environ.get('PINECONE_API_KEY'), environment=os.environ.get('PINECONE_ENV'))
```

```python
# deleting all indexes
indexes = pinecone.list_indexes()
for i in indexes:
  print(f'Deleting all indexes ...', end='')
  pinecone.delete_index(i)
  print('Done')
```

```python
index_name = 'thoughts-on-portland'
if index_name not in pinecone.list_indexes():
  print(f'Creating index {index_name} ... ')
  pinecone.create_index(index_name, dimension=1536, metric='cosine')
  print('Done')
```

```python
vector_store = Pinecone.from_documents(chunks, embeddings, index_name=index_name)
```

**Asking Questions (Similarity Search)**

```python
query = 'where should we eat?'
result = vector_store.similarity_search(query)
print(result)
```
